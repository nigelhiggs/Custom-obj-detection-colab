```
#were gonna try and use time model to clear the output
#by threading are code
from google.colab import output
import time
import threading

#holds a bunch of numbers representing seconds.
times = []
# 14,400 = 4 hrs 
for y in range(14400):
  #for every 300 = 5mins 
  if y %300==0:
    #add number to array
    times.append(y)
  else:
    continue

#function that holds the command to clear output.
def gfg():
    output.clear()
#the length of values stored in array time
for x in range(len(times)):
  #set variable timer with the function to clear output every 5mins
    timer = threading.Timer(times[x],gfg)
    #than execute this command.
    timer.start()


# use the line below to train a fresh model
#!./darknet detector train /content/Custom-obj-detection-colab/obj.data.txt /content/darknet/build/darknet/x64/cfg/yolov3.cfg /content/darknet/darknet53.conv.74 -dont_show

# use the line below to retrain your previous saved weight
!./darknet detector train /content/Custom-obj-detection-colab/obj.data.txt /content/darknet/build/darknet/x64/cfg/yolov3.cfg /content/darknet/build/darknet/x64/backup/yolov3_last.weights /content/darknet/darknet53.conv.74 -dont_show
```

```

def display_image(file_path = '/content/darknet/predictions.jpg'):
    import cv2
    import matplotlib.pyplot as plt
    import os.path

    if os.path.exists(file_path):
      img = cv2.imread(file_path)
      show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
      plt.imshow(show_img)
    else:
      print('failed to open file')
    
display_image()
```
